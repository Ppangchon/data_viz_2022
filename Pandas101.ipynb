Quiz (ให้ใช้เฉพาะที่อาจารย์สอนไปแล้วในวิชานี้)

1.   คำนวณ อายุเฉลี่ย ของ**ผู้ชาย** ในข้อมูลทั้งหมด
2.   คำนวณ อายุเฉลี่ย ของ**ผู้ชาย** ในจังหวัดขอนแก่น
3. หาจำนวนผู้ป่วยที่เป็นคน "**ที่ไม่ใช่คนไทย**"

````python
this_data = data_covid[['sex','age','province_of_onset']] 
````

````python
this_data  #ตัดออกมาจากตารางทั้งหมด จะได้ 3 คอลัมน์
````

````python
male = this_data [this_data ['sex']=='ชาย']
male
````

````python
male['age']
````

````python
sum = 0 #เริ่มที่ 0
N = 0 #จะนับว่ามีจำนวนคนทั้งหมดกี่คน
for a in male['age']:
    
    if a > 0: 
        sum += a # sum = sum + a 
        N += 1

print(f'อายุเฉลี่ย ของ ผู้ชาย {sum/N}')
````

````python
male_KK = male[male['province_of_onset']=='ขอนแก่น'] 

sum = 0
N = 0
for a in male_KK['age']:
    
    if a > 0:
        sum += a # sum = sum + a
        N += 1

print(f'อายุเฉลี่ย ของผู้ชาย ในจังหวัดขอนแก่น {sum/N}')
````

````python
patient = data_covid[data_covid ['nationality'] != 'Thailand'] #กำหนดค่าตัวแปร patient โดยใช้คอลัมน์ nationality ต้องการหาผู้ป่วยที่ "ไม่ใช่คนไทย"
total = patient.count() #ใช้ count ในการหาผลรวมของผู้ป่วยทั้งหมดที่ไม่ใช่คนไทย
total
````
