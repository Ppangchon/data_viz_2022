## เลือกข้อมูล

```python
alldb_1 = alldb[['user','RowKey','date','phone','site','time','urine','ov','PartitionKey']]
alldb_1
```

## Prepare alldb  Clean Missing


```python
alldb_1.isnull().any()  #ขึ้น True แสดงว่ายังมีค่า Missimg
```

```python
alldb_2 = alldb_1.dropna() #ลบข้อมูล missing ทิ้ง
alldb_2
```

```python
alldb_2.isnull().any() #ตรวจสอบค่า mossing อีกครั้งหลัง ลบ missing ทิ้ง
```

## แก้เงื่อนไข user เป็นตัวเลข 5 หลัก

```python
alldb_2 = alldb_2[(alldb_2['user'].astype(str).str.isdigit()) & (alldb_2['user'].str.len() == 5)]   # กำหนดให้ข้อมูล user ที่เป็นตัวเลข & มีตัวเลข เท่ากับ 5 หลัก (ตัวเลขเท่านั้น)
print(alldb_2.shape)    # ตรวจสอบขนาดของข้อมูล                                                                        
alldb_2 # print data ดูว่ามันหน้าตาเป็นยังไง
```

## แก้เงื่อนไข RowKey เป็น String 


```python
set(alldb_2['RowKey'])
```

```python
alldb_2['RowKey'] = alldb_2['RowKey'].astype('string')
alldb_2
```

```python
print(alldb_2.dtypes)
```

##แก้เงื่อนไข เปลี่ยน date เป็น YYYY-MM-DD

```python
import datetime # ทำการ import package datetime 
alldb_2 = pd.DataFrame(alldb_2) 
alldb_2['date'] = pd.to_datetime(alldb_2['date']) # format = "%d/%m/%Y" 
print(alldb_2.shape)
alldb_2
```

##แก้เงื่อนไข เปลี่ยน phone เป็น String

```python
alldb_2['phone'] = alldb_2['phone'].astype('str')
print(alldb_2.shape)
alldb_2
```

### แก้เงื่อนไข เปลี่ยน site เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น)

```python
alldb_2 = alldb_2[(alldb_2['site'].astype(str).str.isdigit()) & (alldb_2['site'].str.len() == 5)]      # เลือกมาเฉพาะ คอลัมน์ site ที่เป็นตัวเลข & มีตัวเลข เท่ากับ 5 ตัว
print(alldb_2.shape)                                                                            
alldb_2
```

### แก้เงื่อนไขเปลี่ยน time เป็น hh-mm-ss

```python
alldb_2['time'] = alldb_2['time'].replace(':','-',regex=True) # กำหนดรูปแบบของ time เป็น hh-mm-ss
alldb_2
```

### แก้เงื่อนไขเปลี่ยน urine มีค่า String

```python
alldb_2['urine'] = alldb_2['urine'].astype('str')
print(alldb_2.shape)
alldb_2
```

###ov เป็นตัวเลข มีค่าตั้งแต่ 0-4 (0 คือไม่พบโรค 1-4 คือพบโรคที่ระดับความรุนแรงต่างๆ)
โดยจะกำหนดให้แทนด้วยตัวเลขดังต่อไปนี้ 
* ติดเชื้อ +2 = 4
* มีเม็ดเลือดแดง = 4
* สีฟ้า หรือน้ำเงิน = 3
* สีน้ำนม = 2
* ขุ่น = 1
* มีตะกอน = 1
* ปกติ (เหลืองใส) = 0
* ใสเป็นน้ำเปล่า = 0

```python
set(alldb_2['ov'])  #ดูข้อมูลคอลัมน์ ov
```

```python
alldb_2['ov'].replace({'ขุ่น':'1','ติดเชื้อ +2 ':'4','ปกติ (เหลืองใส)':'0','มีตะกอน':'1','มีเม็ดเลือดแดง':'4',
'สีน้ำนม':'2','สีฟ้า หรือน้ำเงิน':'3','ใสเป็นน้ำเปล่า':'0',}, inplace=True)  #แทนค่าข้อมูลตามที่เรากำหนด
```

```python
set(alldb_2['ov'])  #เชคข้อมูลคอลัมน์ ov ข้อมูลเป็นตามที่กำหนดแล้ว
```

###PartitionKey เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น)
